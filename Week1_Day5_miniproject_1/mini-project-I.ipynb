## Task
What are the different modes of transport which are operated by Transfer for London? How many of modes do they have?

Print the list with different modes of transport, plus their count. Example output:
```
[bus, cable-car,.....]
Number of different modes of transport is: xyz
```

We need to search the documentation for correct request.

````python
mode=requests.get("https://api.tfl.gov.uk/Line/Meta/Modes")
mode.status_code
````

````python
mode.json()
````

````python
modes=[]
for modetype in mode.json():
    modes.append(modetype['modeName'])
print (modes)
print ("Number of different modes of transport is:",len(modes))
````

## Task
How many BikePoints in London are operated by Transfor for London? How many docks are in **all** BikePoints? There is the information for empty and full docks for each BikePoint.

````python
BikePoints=requests.get("https://api.tfl.gov.uk/BikePoint")
BikePoints.json()
````

````python
total_BP=len(BikePoints.json())
print("Total number of BikePoints in London are:",total_BP)
````

````python
list1=[]
for dock in BikePoints.json():
    for i in dock['additionalProperties']:
        if i['key']=='NbDocks':
            list1.append(int(i['value']))
        
print('Number of docks in all bikepoints',sum(list1))
            
    
````

## Task
How many station has `victoria` line?

````python
vic_stations=requests.get('https://api.tfl.gov.uk/Line/VICTORIA/StopPoints')
vic_stations.json()
````

````python
numberofstations=0
for i in vic_stations.json():
    for j in i['lines']:
        if j['id']=='victoria':
            numberofstations+=1
print('Victoria line has',numberofstations)
````

## Task
Plan the journey from Heathrow Airport to Tower Bridge using Bus and Tube? Which way is faster? Example output:
```
Planned duration:
Bus: x minutes
Tube: y minutes
```

We need to search the documentation for correct requests and parameters we need.

````python
import datetime as datetime
from pytz import timezone
````

````python
timezone(("Europe/London"))
timeinLondon= datetime.datetime.now(timezone("Europe/London")).isoformat()
print(timeinLondon)
date=str('?date='+timeinLondon[:4]+timeinLondon[5:7]+timeinLondon[8:10])
time=str('&time='+timeinLondon[11:13]+timeinLondon[14:16])
````

````python
Heathrow_coordinates='51.4700, -0.4543'
Tower_Bridge_coordinates='51.5055, -0.0754'
````

````python
Bus_time=requests.get(f'https://api.tfl.gov.uk/Journey/JourneyResults/{Heathrow_coordinates}/to/{Tower_Bridge_coordinates}{date}{time}'+'&timeIs=Departing'+'&mode=bus')
Tube_time=requests.get(f'https://api.tfl.gov.uk/Journey/JourneyResults/{Heathrow_coordinates}/to/{Tower_Bridge_coordinates}{date}{time}'+'&timeIs=Departing'+'&mode=tube')
````

````python
Bus_time.json()
````

````python
Tube_time.json()
````

````python
x_minutes=str(Bus_time.json()["journeys"][0]['duration'])
y_minutes=str(Tube_time.json()["journeys"][0]['duration'])
print('Planned Duration:')
print('Bus:',x_minutes,'minutes')
print('Tube:',y_minutes,'minutes')
````
