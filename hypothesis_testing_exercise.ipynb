### Task 2
Test the hypothesis that the offensive productions stats of the Cleveland Cavaliers and Golden State Warriors (the teams that met in the finals that year) were from the same distribution in the 2015/2016 season.

Offensive production refers to two variables: **PTS (Points)** and **FG_PCT (Field Goal Percentage)**. We will need to do two separate hypothesis tests, one for each variable.

```python
gsw15=nba15[(nba15['TEAM_NAME'] == 'Golden State Warriors')]
cc15=nba15[(nba15['TEAM_NAME'] == 'Cleveland Cavaliers')]
gsw15.head()

```

```python
# Null Hypothesis PTS(CC15) = PTS(GSW15)
cc15pts = cc15['PTS'].tolist()
gsw15pts = gsw15['PTS'].tolist()
ptest= st.stats.ttest_ind(cc15pts,gsw15pts)
print (ptest)
if ptest[1] > 0.05:
    print('accept null hypothesis')
elif ptest[1] < 0.05:
    print('reject null hypothesis')
```

```python
# Null Hypothesis FG_PCT(CC15) = FG_PCT(GSW15)
cc15fg = cc15['FG_PCT'].tolist()
gsw15fg = gsw15['FG_PCT'].tolist()
ptest= st.stats.ttest_ind(cc15fg,gsw15fg)
print (ptest)
if ptest[1] > 0.05:
    print('accept null hypothesis')
elif ptest[1] < 0.05:
    print('reject null hypothesis')
```

### Task 3
Test the hypothesis that the number of points (PTS) scored by Cleveland Cavaliers changed significantly after the head coach changed in the 2015/2016 season.

- **Coach Blatt was fired on 24th of Jan, 2016**. 

Use the data from seasons 2014/2015 and 2015/2016 - those are years when Cleveland was coached by Blatt. 

**We have two possible solutions to try here:**
- Take the same amount of games from before and after and try t-test.
- Take all the games from before and after and look for the right test to compare two samples with different sizes. (You will need to go through the scipy documentation or google to figure out what kind of test is required.)

```python
# Null Hypothesis (PTS beforeJan24,2016) > (PTS after Jan24,2016)
cc14_pts=nba14[(nba14['TEAM_NAME']== 'Cleveland Cavaliers')]
pts_14=cc14_pts['PTS'].tolist()

#Coach Blatt before fire in 2015, split year in 2 halves
cc15_pts_blatt=nba15[(nba15['TEAM_NAME'] == 'Cleveland Cavaliers')]
cc_blatt=cc15_pts_blatt[cc15_pts_blatt['GAME_DATE']<'2016-01-24']
pts_15_blatt = cc_blatt['PTS'].tolist()

#Coach blatt is fired, remainder season points after date to list
cc_noblatt=cc15_pts_blatt[cc15_pts_blatt['GAME_DATE']>'2016-01-24']
pts_15_noblatt=cc_noblatt['PTS'].tolist()

before= pts_14+pts_15_blatt
after=pts_15_noblatt
```

```python
pvalue=st.stats.ttest_ind(before,after)[1]
if pvalue > 0.05:
    print('accept null hypothesis, points changed significantly after firing blatt')
elif pvalue < 0.05:
    print('reject Null Hypothesis, points did not change after firing blatt')
```

Test the hypothesis that **number of blocks (BLK)** are from the same distribution in both the NBA playoffs and in the NBA regular season for 2015/2016 seaon for the **Toronto Raptors**. 

- We will be working with two samples with different sizes again.

```python
# Null Hypothesis: Number of BLK in season = playoffs for Toronto Raptors
tt15_reg=nba15[(nba15['TEAM_NAME']== 'Toronto Raptors')]
tt15_regblks=tt15_reg['BLK'].tolist()

tt15_playoffs=playoffs16[(playoffs16['TEAM_NAME']=='Toronto Raptors')]
tt15_playoffsblks=tt15_playoffs['BLK'].tolist()

pvalue=st.stats.ttest_ind(tt15_regblks,tt15_playoffsblks)[1]
if pvalue > 0.05:
    print('Accept null hypothesis, blocks are same')
elif pvalue < 0.05:
    print('Reject null hypothesis, blocks are diff')
```

### Task 6
Test the hypothesis that the number of points (PTS) scored by Cleveland Cavaliers is equally distributed for all 3 seasons. 

- In this case, we need a hypothesis test that compares more than 2 distributions at the same. (You will need to go through the scipy documentation or google to figure out what kind of test is required.)

```python
cc13=nba13[(nba13['TEAM_NAME']=="Cleveland Cavaliers")]['PTS'].tolist()
cc14=nba14[(nba14['TEAM_NAME']=="Cleveland Cavaliers")]['PTS'].tolist()
cc15=nba15[(nba15['TEAM_NAME']=="Cleveland Cavaliers")]['PTS'].tolist()
```

```python
pvalue= st.stats.f_oneway(cc13,cc14,cc15)[1]
if pvalue > 0.05:
    print("accept null hypothesis, is equally distributed")
else:
    print("reject null hypothesis, points are not equally distributed")
```
